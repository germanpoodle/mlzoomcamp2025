## Q7. Finding sum of weights



1.   Select all cars from `asia`








```python
df_asia = df[df['origin'] == 'asia']
```

### Select only columns `vehicle_weight` && `model_year` & select first 7 values


```python
df_asia = df_asia[['vehicle_weight', 'model_year']]
df_asia.head(7)
```

```python
# get the underlying numpy array
x = np.array(df_asia[:7])
```

```python
print("Shape of x:", x.shape)
print("Data type of x:", x.dtype)
print("First 5 rows of x:\n", x[:5])
```

```python
x_transpose = x.T
print("Transpose of x:\n", x_transpose)
print("Shape of transpose of x:\n", x_transpose.shape)
print("First 5 rows of transposed x:\n", x_transpose[:5])

# Matrix multiplication of x_transpose and x
XTX = x_transpose @ x
print("\nMatrix multiplication (x_transpose @ x):\n", XTX)
```

```python
XTX_inv = np.linalg.inv(XTX)
print("Inverse of XTX:\n", XTX_inv)
print("Shape of inverse of XTX:\n", XTX_inv.shape)
```

```python
y = np.array([1100, 1300, 800, 900, 1000, 1100, 1200])
y.shape
```

```python
w = XTX_inv.dot(x_transpose).dot(y)
print("Weights (w):\n", w)
print("Shape of weights (w):\n", w.shape)
```
